```python
# Q1. Given a list of integers, write a function to return the sum of prime numbers in that list
list1 = [5,10,15,20,3,11]
def sum_of_prime(n):
    sum_of_prime_numbers=0
    for i in range(0,len(n)):
        flag=0
        num=n[i]
        for j in range(2, (num//2)+1):
            if num%j==0:
                flag=flag+1
                break   
        if (flag==0 and num!=1):
            sum_of_prime_numbers=sum_of_prime_numbers+num
    return sum_of_prime_numbers

print(sum_of_prime(list1))
```

```python
# Q2. Given a list of integers, write a function whether a given list is strictly increasing or not
list1 = [1, 4, 5, 7, 8, 10]
def strictly_increasing(list1):
    res = all(i < j for i, j in zip(list1, list1[1:]))
    return res
print(strictly_increasing(list1))
```

```python
# Q3. Write a function to check whether a given list is expanding or not(the difference between adjacent elements 
# should keep on increasing)
list1 = [4,5,6,3]
def expanding_not(list1):
    diff_list=[]
    for x, y in zip(list1[0::], list1[1::]):
        diff_list.append(y-x)
        
    sort_list= diff_list[:]
    sort_list.sort()
    if(sort_list==diff_list):
        return True 
    else:
        return False
    
expanding_not(list1)
```

```python
# Q4. Write a function to calculate all permuations of a given string (Without usig itertools)
def generatePermutation(string,s,e):  
    c = 0;  
    if(s == e-1):  
        print(string);  
    else:   
        for c in range(s,e):   
            x = list(string);  
            temp = x[s];  
            x[s] = x[c];  
            x[c] = temp;  
            generatePermutation("".join(x),s+1,e);  
            temp = x[s];  
            x[s] = x[c];  
            x[c] = temp;  
  
str = "ABC"  
n = len(str);  
print("All possible permutations are");  
generatePermutation(str,0,n) 
```
